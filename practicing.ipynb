Class Conta

```python
class Conta:
    def __init__(self, titular: str, numero: str, saldo=0):
        self.titular = titular
        self.numero = numero
        self.saldo = saldo
    
    def depositar(self, valor: float):
        format_valor = round(valor, 2)
        
        if format_valor > 0: 
            self.saldo += format_valor
        else: 
            raise Exception("valor de depósito inválido")

    def sacar(self, valor: float):
        format_valor = round(valor, 2)

        if format_valor > 0 and format_valor < self.saldo: 
            self.saldo -= format_valor
        else: 
            raise Exception("valor de saque inválido")

    def get_saldo(self):
        return round(self.saldo, 2)

    def __str__(self):
        return f"Conta {self.numero} de {self.titular} - Saldo: R${round(self.saldo, 2)}"
    

class ContaPoupanca(Conta):
    def __init__(self, titular: str, numero: str, saldo=0, lim_cheque_especial: float):
        super().__init__(titular, numero, saldo)
        self.lim_cheque_especial = lim_cheque_especial

    


```
