Class Conta

```python
class Conta:
    def __init__(self, titular: str, numero: str, saldo: float=0):
        self.titular = titular
        self.numero = numero

        if saldo >= 0:
            self._saldo = round(saldo, 2)
        else:
            raise Exception("saldo inválido")
    
    @property
    def agency(self):
        return f"{self.numero}-001"

    # @agency.setter
    # def agency(self, numero)



    def depositar(self, valor: float):
        format_valor = round(valor, 2)
        
        if format_valor > 0: 
            self._saldo += format_valor
        else: 
            raise Exception("valor de depósito inválido")

    def sacar(self, valor: float):
        format_valor = round(valor, 2)

        if format_valor > 0 and format_valor < self._saldo: 
            self._saldo -= format_valor
        else: 
            raise Exception("valor de saque inválido")

    def get_saldo(self):
        return round(self._saldo, 2)

    def __str__(self):
        return f"Conta {self.numero} de {self.titular} - Saldo: R${round(self._saldo, 2)}"
    

class ContaPoupanca(Conta):
    taxa_juros = 0.005

    def __init__(self, titular: str, numero: str, saldo: float=0):
        super().__init__(titular, numero, round(saldo, 2))

    def aplicar_juros(self):
        self._saldo = self.taxa_juros * self._saldo

    def __str__(self):
        return f"Conta {self.numero} de {self.titular} - Saldo: R${round(self._saldo, 2)} - Juros: {self.taxa_juros * 100}%"
    
    @classmethod
    def mudar_juros(cls, amnt):
        if 0 < amnt < 1:
            cls.taxa_juros = amnt
        else:
            raise Exception("Juros Inválidos")

class ContaCorrente(Conta):
    def __init__(self, titular: str, numero: str, saldo: float=0, lim_cheque_especial: float=0):
        super().__init__(titular, numero, round(saldo, 2))
        
        if(lim_cheque_especial > 0):
            self.lim_cheque_especial = lim_cheque_especial
        else:
            raise Exception("Limite Inválido")

    def sacar(self, valor: float):
        format_valor = round(valor, 2)

        if self._saldo - format_valor > -500:
            self._saldo -= format_valor
        else:
            raise Exception("valor de saque inválido")
    
class Banco:
    num_bancos = 0

    def __init__(self, nome: str, contas: list=[]):
        self.nome = nome
        self.contas = contas
        Banco.num_bancos += 1


    def abrir_conta(self, conta: Conta):
        self.contas.append(conta)

    def listar_contas(self):
        for conta in self.contas:
            print(conta.__str__())
```
